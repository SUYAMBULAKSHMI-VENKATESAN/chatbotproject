```python
from nltk.corpus.reader import wordnet
import nltk
nltk.download('wordnet')

df = pd.read_csv("dialogs.txt", sep='\t', header=None, names=["text"])


# Get the text from the DataFrame by header name
text_by_header_name = df["text"].to_string()


stopwords = set(STOP_WORDS)

def removing_stopwords(text):
  doc = text.split()
  no_stop_words =[word for word in doc if word not in stopwords]
  return " ".join(no_stop_words)

df["text"] = df["text"].apply(removing_stopwords)

display(df["text"])
```
