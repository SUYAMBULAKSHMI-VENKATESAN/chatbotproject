```python

doc = nlp("what the hell are you doing man , and what the hell is wrong with you . who are you man to tell this to me?")

for token in doc:
  if token.is_stop:
    print(token)
```

```python
def removing_stopwords(text):
  doc = nlp(text) #doc variable is used to store the given data
  no_stop_words =[token.text for token in doc if not token.is_stop and not token.is_punct]
  return no_stop_words
```

```python

df["Column_2_new"] = df["Column_2"].apply(removing_stopwords)
df["Column_1_new"] = df["Column_1"].apply(removing_stopwords)
#df.head()
value = display(df.head(3725))
```
