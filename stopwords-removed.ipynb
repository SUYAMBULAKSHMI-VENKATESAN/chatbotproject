```python
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
len(STOP_WORDS)

nlp = spacy.load("en_core_web_sm")
doc = nlp("what the hell are you doing man , and what the hell is wrong with you . who are you man to tell this to me?")

for token in doc:
  if token.is_stop:
    print(token)
```

```python
def removing_stopwords(text):
  doc = nlp(text)
  no_stop_words =[token.text for token in doc if not token.is_stop and not token.is_punct]
  return no_stop_words

removing_stopwords("what the hell are you doing man , and what the hell is wrong with you . who are you man to tell this to me?")
```
