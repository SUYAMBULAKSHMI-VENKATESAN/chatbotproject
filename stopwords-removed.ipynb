```python
from string import punctuation
df = pd.read_csv("dialogs.txt", sep='\t', header=None, names=["column1", "column2"])

# Get the text from the DataFrame by header name
text_by_header_name = df["column1"].to_string() + df["column2"].to_string()


stopwords = set(STOP_WORDS)

def removing_stopwords(text):
  doc = text.split()

  no_stop_words =[word for word in doc if word not in STOP_WORDS and word not in punctuation]
  return " ".join(no_stop_words)

df["column1"] = df["column1"].apply(removing_stopwords)
df["column2"] = df["column2"].apply(removing_stopwords)


display(df["column1"])
display(df["column2"])
```
