```python
import pandas as pd

def encode_categorical_variables(df):
    """Encodes categorical variables using one-hot encoding."""

    categorical_variables = df.select_dtypes(include=['object'])
    for variable in categorical_variables:
        df = pd.get_dummies(df, columns=[variable])

    return df

def convert_numerical_variables_to_float(df):
    """Converts numerical variables to floating-point numbers."""

    numerical_variables = df.select_dtypes(exclude=['object'])
    for variable in numerical_variables:
        df[variable] = df[variable].astype(np.float32)

    return df

```

```python
df = pd.read_csv("cleaned_dailogs.txt", sep='\t', header=None)
#df = pd.read_csv("cleaned_dailogs.txt")

# Encode the categorical variables
df = encode_categorical_variables(df)

# Convert the numerical variables to floating-point numbers
df = convert_numerical_variables_to_float(df)
df
```

```python
import tensorflow as tf
import numpy as np

# Load the training data
train_x = np.loadtxt("train_x.csv", delimiter=",")
train_y = np.loadtxt("train_y.csv", delimiter=",")

# Define the chatbot model
class ChatbotModel(tf.keras.Model):
    def __init__(self, vocab_size, embedding_dim, hidden_dim):
        super(ChatbotModel, self).__init__()

        self.embedding_layer = tf.keras.layers.Embedding(vocab_size, embedding_dim)
        self.encoder = tf.keras.layers.LSTM(hidden_dim)
        self.decoder = tf.keras.layers.LSTM(hidden_dim)
        self.output_layer = tf.keras.layers.Dense(vocab_size)

    def call(self, inputs):
        embeddings = self.embedding_layer(inputs)
        encoder_outputs, encoder_state = self.encoder(embeddings)
        decoder_outputs, _ = self.decoder(encoder_outputs, initial_state=encoder_state)
        outputs = self.output_layer(decoder_outputs)

        return outputs

# Instantiate the chatbot model
model = ChatbotModel(vocab_size=len(train_x[0]), embedding_dim=128, hidden_dim=256)

# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(train_x, train_y, epochs=100)

# Save the model
model.save("chatbot_model.h5")
```
